# Normalización de Fechas en HomicidioIntencional

Este notebook carga los archivos de homicidios y convierte la columna `FECHA HECHO` del formato `yyyymmdd` al formato `yyyy/mm/dd`.

Luego guarda los cambios en los archivos Excel originales.

```python
import pandas as pd
import glob
import os
from pathlib import Path
```

```python
def convertir_fecha_hecho(valor):
    """
    Convierte fechas de formato yyyymmdd a yyyy/mm/dd
    
    Args:
        valor: Valor en formato yyyymmdd (puede ser int, float o str)
    
    Returns:
        String en formato yyyy/mm/dd o NaN si no se puede convertir
    """
    if pd.isna(valor):
        return None
    
    # Convertir a string y remover espacios
    valor_str = str(valor).strip()
    
    # Si es 'nan', retornar None
    if valor_str.lower() == 'nan':
        return None
    
    # Si tiene menos de 8 caracteres, no es válido
    if len(valor_str) < 8:
        return valor_str
    
    # Extraer partes: yyyy(0-4), mm(4-6), dd(6-8)
    try:
        yyyy = valor_str[:4]
        mm = valor_str[4:6]
        dd = valor_str[6:8]
        
        # Validar que sean números
        if yyyy.isdigit() and mm.isdigit() and dd.isdigit():
            # Validar rango de meses y días
            mm_int = int(mm)
            dd_int = int(dd)
            if 1 <= mm_int <= 12 and 1 <= dd_int <= 31:
                return f"{yyyy}/{mm}/{dd}"
        
        return valor_str
    except Exception as e:
        print(f"Error procesando valor {valor}: {e}")
        return valor_str
```

```python
# Obtener lista de archivos Excel
root = 'bd-homicidios'
files = sorted(glob.glob(os.path.join(root, 'HomicidioIntencional*.xlsx')))

print(f"Se encontraron {len(files)} archivos Excel:")
for f in files:
    print(f"  - {f}")
```

```python
# Procesar cada archivo
archivos_procesados = []

for filepath in files:
    print(f"\nProcesando: {filepath}")
    
    try:
        # Leer el archivo Excel
        df = pd.read_excel(filepath)
        print(f"  Filas: {len(df)}")
        
        # Verificar que existe la columna FECHA HECHO
        if 'FECHA HECHO' not in df.columns:
            print(f"  ⚠️  Advertencia: Columna 'FECHA HECHO' no encontrada. Columnas disponibles: {list(df.columns)}")
            continue
        
        # Guardar valores originales para comparación
        original_count = df['FECHA HECHO'].notna().sum()
        
        # Aplicar la función de conversión
        df['FECHA HECHO'] = df['FECHA HECHO'].apply(convertir_fecha_hecho)
        
        converted_count = df['FECHA HECHO'].notna().sum()
        
        # Guardar el archivo modificado
        df.to_excel(filepath, index=False)
        print(f"  ✓ Archivo guardado exitosamente")
        print(f"    - Valores originales no nulos: {original_count}")
        print(f"    - Valores convertidos no nulos: {converted_count}")
        
        # Mostrar algunos ejemplos
        print(f"    - Ejemplos (primeros 5):")
        for idx, val in enumerate(df['FECHA HECHO'].head(5).values):
            print(f"      {idx+1}. {val}")
        
        archivos_procesados.append(filepath)
        
    except Exception as e:
        print(f"  ❌ Error: {e}")

print(f"\n{'='*50}")
print(f"Resumen: {len(archivos_procesados)} archivos procesados exitosamente")
print(f"{'='*50}")
```

```python
# Verificación final: cargar uno de los archivos y mostrar la columna FECHA HECHO
print("\nVerificación de los cambios (primer archivo):")
if archivos_procesados:
    df_check = pd.read_excel(archivos_procesados[0])
    print(f"\nArchivo: {archivos_procesados[0]}")
    print(f"\nPrimeras 10 fechas convertidas:")
    print(df_check[['FECHA HECHO']].head(10).to_string(index=False))
    print(f"\n✓ Todas las fechas ahora están en formato yyyy/mm/dd")
```
