Objetivo: La idea por la que se decidió analizar los datos de homicidios intencionales a nivel nacional (Colombia) era entender la realidad social y política del país. El rango de estudio (2015-2025) abarca tres periodos presidenciales distintos y la implementación de diversas estrategias de seguridad. Los datos permiten evaluar, de manera objetiva y sin sesgos ideológicos, qué políticas han tenido un efecto real en la preservación de la vida.

Identificación de patrones y tendencias: Colombia ha vivido cambios significativos en su dinámica de seguridad tras el Acuerdo de Paz de 2016 y acuerdos de Paz Total.

Comprensión de Factores de Riesgo: Cruzar los datos de homicidios podrá a ayudar a  entender qué factores están impulsando la violencia en regiones específicas.

- Total de homicidios en Colombia entre el 2015 y el 2025
- Cual es el departamento con más homicidios
- En que fechas se presentan la mayor cantidad de homicidios
- Cual es el arma más común para efectuar estos homicidios
- Cual es la población más afectada, Hombres, Mujeres o adolecentes y niños.
- Como le fue a Antioquia (Por municio)
- Cual es la población más afectada en Antioquia, Adultos, adolecentes o niños.
